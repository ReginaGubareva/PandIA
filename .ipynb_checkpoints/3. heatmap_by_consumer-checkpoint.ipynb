```python
frames = [a_2013, a_2014, a_2016, a_2017, a_2018, a_2019, a_2020]
df = pd.DataFrame(pd.concat(frames), dtype=int)
print(len(df.index))
df.sample(5)
```

# Prepare data

```python
by_ano = df.groupby(['Ano', 'Mes', 'Cons_numero'], as_index=False).aggregate({'Consumo_abs':'sum'})
by_ano.head()
```

Choose 50 random consumidores with numbers from 0 to 2132971

```python
# choose the 50 random consumidores and put into results_50
random_consumidor = df['Cons_numero'].sample(50)
```

```python
by_ano = by_ano.groupby('Ano')
```

```python
years_list = [2013, 2014, 2016, 2017, 2018, 2019, 2020]
```

```python
random_consumidor.head()
```

# Draw with fix year 

```python
# Try for one ano
ano_2020 = by_ano.get_group(2020).drop(columns=['Ano'])
for x in random_consumidor:
    current = ano_2020.loc[ano_2020['Cons_numero'] == x].drop(columns=['Cons_numero'])
    current['Mes'] = current['Mes'].apply(lambda x: calendar.month_abbr[x])
    sns.lineplot(data=current, x="Mes", y="Consumo_abs").set_title(2020)
```

```python
fig, axes = plt.subplots(ncols=1, nrows=7, figsize=(15, 55))
for i, ax in zip(years_list, axes.flat):
    ano = by_ano.get_group(i).drop(columns=['Ano'])
    for x in random_consumidor:
        current = ano.loc[ano['Cons_numero'] == x].drop(columns=['Cons_numero'])
        current['Mes'] = current['Mes'].apply(lambda x: calendar.month_abbr[x])
        sns.lineplot(data=current, x="Mes", y="Consumo_abs", ax = ax).set_title(2020)
plt.show()
```
